```python
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
import numpy as np
import pylab as pl
%matplotlib inline
from matplotlib import collections  as mc

def circle(mod, mult):
    # represent a line segment as a list of pairs:
    # [ (x1, y1), (x2, y2)]
    lines = []
    for n in range(mod):
        angle1 = 2 * np.pi / mod * n
        angle2 = mult * angle1
        lines.append([ (np.cos(angle1), np.sin(angle1)), (np.cos(angle2), np.sin(angle2)) ])
    #print(lines)
    # graph line segments with matplotlib

    lc = mc.LineCollection(lines, colors = [(1, 0, 1, 1)], linewidths=0.6)
    fig, ax = pl.subplots()
    fig.set_size_inches(11.5, 11.5)
    ax.add_collection(lc)
    ax.autoscale()
    ax.margins(0.1)

interactive_plot = interactive(circle, {'manual': True}, mod=(0, 500), mult=(0, 100))
interactive_plot

circle(400, 60)
```
