```python
import iexfinance as iex
from datetime import datetime

start = datetime(2017, 1, 1)
end = datetime(2017, 12, 31)

aapl_dataframe = iex.get_historical_data("AAPL", start, end, output_format = "pandas")
```

Do something interesting with the data. Do the returns look normally distributed?

1. Extract the adjusted close prices. 
2. Use numpy's `log` and pandas `shift` functions to compute returns. 
3. Use matplotlib.pyplot and scipy.stats to create a histogram and probability plot for AAPL stock returns.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import numpy as np

aapl_close_series = aapl_dataframe['close']
aapl_close_series

def returns(time_series):
    return np.log((time_series / time_series.shift(1))[1:])

r = returns(aapl_close_series)
r

def plot_hist_pp(returns_time_series):
    fig, (histo, pp) = plt.subplots(1, 2, figsize=(12, 4))
    # create a histogram with 20 bins, a title, and labeled axes
    
    histo.hist(returns_time_series, 50, 
               facecolor = 'g', 
               alpha = 0.75)
    
    histo.set_xlabel('log return')
    
    # create a probability plot 
    stats.probplot(returns_time_series, plot = plt)
    plt.show()
    
plot_hist_pp(r)
print(np.std(r))
print(stats.skew(r)) # positive and larger than 0, so skew right
print(stats.kurtosis(r)) # Bigger than 3, which breaks normality assumption
```

```python
import quandl

start = datetime(2017, 1, 1)
end = datetime(2017, 12, 31)

eur_usd_df = quandl.get('CME/EUSM2018', start_date = start, end_date = end)
plot_hist_pp(returns(eur_usd_df['Settle']))
```
